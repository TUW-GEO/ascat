First we import the necessary class

```python
import os
from ascat.cgls import SWI_TS
```

By default we should have the grid file from the SWI-STATIC collection and the unzipped SWI-TS products in one folder like so:

```python
ls tests/test-data/sat/cglops/swi_ts
```

Now we can initialize the data reader

```python
data_path = os.path.join('tests', 'test-data', 'sat', 'cglops', 'swi_ts')
rd = SWI_TS(data_path)
```

With this object we can read the data

```python
data = rd.read_ts(3002621)
print(data)
```

Since the returned value is a pandas.DataFrame we can plot the data easily.

```python
import matplotlib.pyplot as plt
%matplotlib inline
data[['SWI_001', 'SWI_010']].plot()
plt.show()
```
